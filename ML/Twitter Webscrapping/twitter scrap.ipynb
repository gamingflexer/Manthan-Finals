```python
# USING API

auth=tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_secret)
```

```python
api=tweepy.API(auth)
```

```python
t=[]
scrap=tweepy.Cursor(api.user_timeline, id='raqqa_mcr',tweet_mode="extended").items(50)
scrap
```

```python
for j in scrap:
    t.append(j.full_text)
    print(t)
```

```python
data = pd.DataFrame(t)
data.columns = ['tweets']
data
```

```python
#cleaning tweets 
```

```python
data['tweets'] = data['tweets'].apply(lambda x: x.lower())
data['tweets'] = data['tweets'].apply(lambda x: " ".join (re.sub(r'([a-zA-Z0-9+._-]+@[a-zA-Z0-9+._-]+\.[a-zA-Z0-9_-]+)'," ",x).split()))
data['tweets'] = data['tweets'].apply(lambda x:" ".join (re.sub(r'(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', "", x).split()))
data['tweets'] = data['tweets'].apply(lambda x:" ".join (re.sub('[^A-Z a-z 0-9-]+', "", x).split()))
data['tweets'] =data['tweets'].apply(lambda x:" ".join (re.sub('rt', "",x).split()))
```
