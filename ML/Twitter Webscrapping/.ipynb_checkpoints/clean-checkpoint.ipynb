```python
import pandas as pd
import re
import twint
```

```python
# USING WEB SCRAPPING 

#scrap 

c = twint.Config()

#for k in ky:
c.Search = "ISIS"
c.Limit= 300
c.Min_likes = 3
c.Custom["tweet"] = ["id", "username","tweet"]
c.Output="10.json"
c.Store_json=True

c.Hide_output = True

twint.run.Search(c)


```

```python
data = pd.read_csv (r'1.csv')
print (data)
```

```python
data['tweet'] = data['tweet'].apply(lambda x: x.lower())
data['tweet'] = data['tweet'].apply(lambda x: " ".join (re.sub(r'([a-zA-Z0-9+._-]+@[a-zA-Z0-9+._-]+\.[a-zA-Z0-9_-]+)'," ",x).split()))
data['tweet'] = data['tweet'].apply(lambda x:" ".join (re.sub(r'(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', "", x).split()))
data['tweet'] = data['tweet'].apply(lambda x:" ".join (re.sub('[^A-Z a-z 0-9-]+', "", x).split()))
data['tweet'] =data['tweet'].apply(lambda x:" ".join (re.sub('rt', "",x).split()))
```
